```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_association_football_stadiums_by_capacity"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
}
```

```python
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
table = soup.select("table", {"class":"wikitable sortable"})[1]
table
```

```python
table_rows = table.find_all("tr")
table_rows
```

```python
len(table_rows)
```

```python
for i in range(1,len(table_rows)):
    tds=table_rows[i].find_all('td')
    print({'Rank' : i,
        'Stadium' : tds[0].text,
        'Seating capacity' : tds[1].text,
        'Region' : tds[2].text,
        'Country' : tds[3].text,
        'City' : tds[4].text,
        'Images' : tds[5].find('img').get('src').split("//")[1] if tds[5].find('img') else "No image available" ,
        'Home_Team' : tds[6].text})
```

```python
def clean_text(text):
    text = str(text).strip()
    if text.find("\xa0"):
        text = text.split("\xa0")[0]

    return text.replace('/n', '')
```

```python
data = []

for i in range(1, len(table_rows)):
    tds = table_rows[i].find_all('td')
    values = {
        'Rank' : i,
        'Stadium' : clean_text(tds[0].text),
        'Seating capacity' : tds[1].text,
        'Region' : tds[2].text,
        'Country' : clean_text(tds[3].text),
        'City' : tds[4].text,
        'Images' : tds[5].find('img').get('src').split("//")[1]  if tds[5].find('img') else "No image available",
        'Home_Team' : tds[6].text
    }

    data.append(values)

print(data) 
```

```python
def get_page(url):
  import requests
  from bs4 import BeautifulSoup

  print(f"Getting page: {url}")

  try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, "html.parser")
    return soup
  except requests.exceptions.RequestException as e:
    print(f"Error: {e}")
    return None
```
